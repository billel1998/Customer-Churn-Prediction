*** Step 1: Separate Features and Target

```python
# Target (y) and features (X)
y = df["Churn"]
X = df.drop("Churn", axis=1)
```

Step 2: Handle Categorical Variables (One-Hot Encoding)

```python
# One-hot encode categorical features
X_encoded = pd.get_dummies(X, drop_first=True)
```

Step 3: Combine Encoded Features with Target

```python
# Final dataset ready for modeling
df_model = pd.concat([X_encoded, y], axis=1)

# Check shape
print(df_model.shape)
df_model.head()
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
# 4) Simple stratified train/test split (recommended baseline)
#    - test_size=0.2 means 80% train / 20% test
#    - stratify=y preserves churn ratio in both sets
#    - random_state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, stratify=y, random_state=42
)

```

```python
# Confirm shapes and class proportions
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
print("Train churn distribution:\n", y_train.value_counts(normalize=True))
print("Test churn distribution:\n", y_test.value_counts(normalize=True))
print()
```

```python
# 6) Scaling numeric features (fit scaler on TRAIN only to avoid leakage)
#    - Tree models do NOT need scaling, but linear models and distance-based models do.
num_cols = ["SeniorCitizen", "tenure", "MonthlyCharges", "TotalCharges"]

# We'll scale X_train/X_test from the 80/20 example
scaler = StandardScaler()
X_train_scaled = X_train.copy()            # safe copies so we don't mutate original
X_test_scaled = X_test.copy()
```

```python
# Fit scaler on training numeric columns then transform both train and test
X_train_scaled[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test_scaled[num_cols] = scaler.transform(X_test[num_cols])
```

```python
# Quick check: mean ~0 and std ~1 on train (for numeric cols)
print("Train scaled means (numeric cols):", X_train_scaled[num_cols].mean().round(3).to_dict())
print("Train scaled stds (numeric cols):", X_train_scaled[num_cols].std().round(3).to_dict())
```

Logistic Regression baseline

LogisticRegression → the model.

classification_report → gives precision, recall, f1, accuracy.

confusion_matrix → helps see false positives/negatives.

roc_auc_score → measures overall separability (how well the model ranks churners vs non-churners).

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

Step 2: Initialize and fit the model

max_iter=1000 → ensures convergence (sometimes default 100 is not enough).

class_weight="balanced" → handles class imbalance (27% churn vs 73% no churn).

random_state=42 → reproducibility.

fit(...) → trains the model on training data.

```python
# Use the scaled features for Logistic Regression
log_reg = LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42)

# Fit model
log_reg.fit(X_train_scaled, y_train)
```

Step 3/ predictions

```python
# Predict on test set
y_pred = log_reg.predict(X_test_scaled)

# Also get probabilities (for ROC-AUC)
y_prob = log_reg.predict_proba(X_test_scaled)[:,1]
```

Step 4: Evaluation

```python
# Confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Classification report (precision, recall, f1)
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# ROC-AUC
roc_auc = roc_auc_score(y_test, y_prob)
print("\nROC-AUC Score:", round(roc_auc, 3))
```

Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Initialize Random Forest (basic, no heavy tuning)
rf = RandomForestClassifier(
    n_estimators=200,        # number of trees (good balance of speed & accuracy)
    class_weight="balanced", # handles imbalance
    random_state=42,
    n_jobs=-1                # use all CPU cores
)

# Fit on training data (use UN-SCALED features!)
rf.fit(X_train, y_train)

# Predictions
y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:,1]

# Evaluation
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))
print("\nROC-AUC Score:", round(roc_auc_score(y_test, y_prob_rf), 3))
```

So Random Forest is more conservative: it only calls someone a churner when it’s more confident, which improves precision but loses recall.

🔑 Business Lens

If the company’s priority is not missing churners → Logistic Regression is better (catch 3 out of 4).

If the company’s priority is fewer false alarms (don’t waste offers on loyal customers) → Random Forest is better.

Ideally, we’d tune the probability threshold of Random Forest to increase recall (instead of sticking to 0.5 default). That way, we get the best of both worlds.
